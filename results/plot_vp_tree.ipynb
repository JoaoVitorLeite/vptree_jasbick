```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import io
import seaborn as sns
from os import listdir
from os.path import isfile, join
%matplotlib inline
```

```python
def readResults(path, group):
    onlyfiles = [join(path,f) for f in listdir(path) if isfile(join(path, f))]
    info_pivots = {}
    query_data = []
    label = []
    
    for path in onlyfiles:
        try:
            file = open(path)
            lines = file.readlines()
            data = [line.rstrip().split(",") for line in lines]
            key = data[0][2]
            data[0].pop(2)
            value = data[0]
            info_pivots[key] = value
            data.pop(0)

            data[0] = ["Pivot"] + data[0]

            for i in range(1, len(data)):
                data[i] = [key] + data[i]

            label = data[0]
            data.pop(0)

            query_data.extend(data)

        finally:
            file.close()
    
    query_data.insert(0, label)
    data = [" ".join(line) for line in query_data]
    df = pd.read_csv(io.StringIO('\n'.join(data)), delim_whitespace=True)
    df.columns = ["Pivot", "K", "Time", "Distance Count"]
    
    if group:
        group_df = df.groupby(["Pivot", "K"])[["Distance Count"]].median().reset_index()
        return group_df
    else:
        return df
```

```python
df = readResults("ulcer_vptree/", True)
```
